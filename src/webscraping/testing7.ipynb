```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('../data/residence.csv',sep=";", index_col=False)  
df.head()
```

```python
#df = df.loc[df['house_zip_code'] == 3400]

#Data
df_x = df[['house_zip_code', 'house_rooms', 'house_square_meters',
       'house_year']]
#Target
df_y = df[['house_price']]

df_x
#df_x
```

```python
reg = LinearRegression()

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.10, random_state=2 )

norm = MinMaxScaler().fit(x_train)


# transform training data
x_train_norm = norm.transform(x_train)


# transform testing dataa
x_test_norm = norm.transform(x_test)
```

```python
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```
